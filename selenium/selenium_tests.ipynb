```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Test 1 - Homepage loads
driver.get("http://localhost/opencart")
assert "Your Store" in driver.title, "Homepage title mismatch"
print(" Test 1 PASSED - Homepage loaded:", driver.title)

# Test 2 - Search for a product
search_box = driver.find_element(By.NAME, "search")
search_box.send_keys("MacBook")
search_box.send_keys(Keys.RETURN)
time.sleep(2)
assert "MacBook" in driver.title or "Search" in driver.title
print(" Test 2 PASSED - Search works")

driver.quit()
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("http://localhost/opencart")
time.sleep(2)

# Navigate to MacBook product page
driver.get("http://localhost/opencart/index.php?route=product/product&product_id=43")
time.sleep(2)

# Click Add to Cart button
add_to_cart = driver.find_element(By.ID, "button-cart")
add_to_cart.click()
time.sleep(2)

# Check success message
success = driver.find_element(By.CSS_SELECTOR, ".alert-success")
assert "Success" in success.text
print("Test 3 PASSED - Product added to cart")
print("Message:", success.text)

driver.quit()
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("http://localhost/opencart/index.php?route=account/register")
time.sleep(2)

# Fill in registration form
driver.find_element(By.ID, "input-firstname").send_keys("Mohamed")
driver.find_element(By.ID, "input-lastname").send_keys("Lokma")
driver.find_element(By.ID, "input-email").send_keys("test88@test.com")
driver.find_element(By.ID, "input-password").send_keys("Test1234!")
time.sleep(1)

# Agree to privacy policy
privacy = driver.find_element(By.NAME, "agree")
driver.execute_script("arguments[0].click();", privacy)
time.sleep(1)

# Click Continue button (the registration submit button)
buttons = driver.find_elements(By.TAG_NAME, "button")
continue_btn = buttons[3]  # Button 3 is Continue
driver.execute_script("arguments[0].click();", continue_btn)
time.sleep(3)

print("Page title:", driver.title)
print("Current URL:", driver.current_url)

# Check success
if "account/success" in driver.current_url or "Your Account" in driver.page_source:
    print(" Test 4 PASSED - User registered successfully")
else:
    print(" Test 4 FAILED")

driver.quit()
```
